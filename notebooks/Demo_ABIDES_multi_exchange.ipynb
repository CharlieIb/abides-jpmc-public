```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
a_mod, b_mod = 2, 5
# Option 2: Sharply peaked
a_sharp, b_sharp = 5, 14.5

a_sharper, b_sharper = 10, 29

# Choose which set to use
a, b = a_sharp, b_sharp # Using the sharp peak for this example

min_val = 5
max_val = 70

# --- Generate a Single Random Value ---
# Draw a random sample from the Beta(a, b) distribution (a value between 0 and 1)
random_sample = beta.rvs(a, b)
scaled_value = min_val + random_sample * (max_val - min_val)

print(f"Alpha: {a}, Beta: {b}")
print(f"Random sample (0-1): {random_sample:.4f}")
print(f"Scaled random value: {scaled_value:.2f}")
```

```python
x = np.linspace(0, 1, 500)
y = beta.pdf(x, a, b)
scaled_x = min_val + x * (max_val - min_val)
mode_location = min_val + ((a - 1) / (a + b - 2)) * (max_val - min_val)

```

```python
plt.figure(figsize=(10, 5))
plt.plot(scaled_x, y, lw=3, label=f'Beta PDF (a={a}, b={b})')
plt.axvline(mode_location, color='r', linestyle='--', label=f'Peak at â‰ˆ {mode_location:.2f}')
plt.title("Shape of the Beta Distribution Scaled to [10, 75]")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
